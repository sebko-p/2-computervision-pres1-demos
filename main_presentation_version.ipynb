```python
import cv2
import matplotlib.pyplot as plt
from mtcnn import MTCNN
import face_recognition
import numpy as np
import time
```

```python
## CONSTANTS ## 
which_picture = 0 # starting from 0
number_of_faces = [1,18,7,7,4,4,21,8,8][which_picture]
path_to_image = ['img/steve1.png', 'img/group1.png', 'img/group2.png', 'img/group3.png', 'img/heads_turned1.jpeg', 'img/heads_turned2.png','img/21people.png',
                     'img\facial_occlusion1.png','img\facial_occlusion1.png']
```

```python
## UTILITY METHODS ##
def measure_execution_time(method, arg):
    start_time = time.time()
    method(arg)
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f"Execution of {method.__name__} took: {elapsed_time:.4f} seconds")

def image_reset():
    return cv2.imread(path_to_image[which_picture])

def show_image(img):
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(5,10))
    plt.imshow(img)
    plt.axis('off')

## METHODS WHICH DETECT FACES FROM THE LIBRARIES ##
def haarcascade(img_haar):
    haarcascade_frontalface_path = 'misc_dep\haarcascade_frontalface_default.xml'
    gray_image = cv2.cvtColor(img_haar, cv2.COLOR_BGR2GRAY)
    face_classifier = cv2.CascadeClassifier(haarcascade_frontalface_path)
    face = face_classifier.detectMultiScale(gray_image)
    for (x, y, w, h) in face:
        cv2.rectangle(img_haar, (x, y), (x + w, y + h), (0, 255, 0), 2)
    return img_haar, len(face)
```
